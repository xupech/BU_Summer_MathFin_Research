```python
"""
Getting data from Yahoo Finance for 100 listed companies

"""


start_date = '2010-01-02'
end_date = '2015-12-31'
tickers = ['AA', 'AAL', 'AMZN', 'AAPL', 'ABT', 'ADM', 'AEM', 'AKS', 'ALL', 'APTV', # 10
              'EXC', 'F', 'STZ', 'EL', 'SPWR', 'KO', 'WMT', 'HD', 'NEE', 'CSX', # 20
              'MCD', 'SPG', 'FSLR', 'NSC', 'HON', 'GOOGL', 'COP', 'MRO', 'C', 'CRM', # 30
              'QCOM', 'GILD', 'CVX', 'FB', 'MRK', 'DE', 'AXP', 'V', 'SPY', 'BA', # 40
              'BAC', 'GS', 'JPM', 'BBY', 'BBBY', 'QQQ', 'BRK-A', 'CAG', 'CE', 'CLB' , # 50
              'CMI', 'JEC', 'MMM', 'DAL', 'DG', 'DHI', 'FCFS', 'ENR', 'FDX', 'NKE', # 60
              'FRO', 'GPS', 'H', 'GWW', 'HES', 'HFC', 'VLO', 'HOG', 'HRB', 'HSY', # 70
              'GE', 'T', 'VZ', 'S', 'UPS', 'JNJ', 'K', 'KBH', 'REV', 'KGC', # 80
              'L', 'LUV', 'LYB', 'AWK', 'PFE', 'PG', 'RCL', 'CCI', 'RIG', 'RTN', # 90
              'SAM', 'SB', 'SBUX', 'SHAK', 'SLB', 'SIX', 'TGT', 'TOL', 'TSN', 'TWTR' # 100
              ]

df = data.DataReader(tickers, 'yahoo', start_date, end_date)
```

```python
print("No. of missing prices before : {}".format(df.isnull().sum().sum()))
working_days=pd.date_range(start=start_date, end=end_date, freq='B')
prices=df.reindex(working_days)
prices=prices.fillna(method='ffill')
print("No. of missing prices after : {}".format(prices.isnull().sum().sum()))
prices.tail()
```
